```python
from eth_abi import encode as encode_abi
from eth_utils import keccak
from web3 import Web3
from web3.middleware.proof_of_authority import ExtraDataToPOAMiddleware

w3 = Web3(Web3.HTTPProvider("https://polygon-rpc.com"))
w3.middleware_onion.inject(ExtraDataToPOAMiddleware, layer=0)

SAFE_B = Web3.to_checksum_address(POLYMARKET_SAFE)
DEST = Web3.to_checksum_address(METAMASK_EOA)

# -- your ERC20 calldata here --
data = data

# Safe-specific values
to = DEST
value = 0
operation = 0
safeTxGas = 0
baseGas = 0
gasPrice = 0
gasToken = Web3.to_checksum_address("0x0000000000000000000000000000000000000000")
refundReceiver = Web3.to_checksum_address("0x0000000000000000000000000000000000000000")

# ❗ Correct Safe nonce
nonce_abi = [
    {
        "inputs": [],
        "name": "nonce",
        "outputs": [{"type": "uint256"}],
        "stateMutability": "view",
        "type": "function",
    }
]
nonce_contract = w3.eth.contract(SAFE_B, abi=nonce_abi)
nonce = nonce_contract.functions.nonce().call()
print("Safe nonce =", nonce)

# === Encode the Safe tx struct ===
packed = encode_abi(
    [
        "address",
        "uint256",
        "bytes",
        "uint8",
        "uint256",
        "uint256",
        "uint256",
        "address",
        "address",
        "uint256",
    ],
    [
        to,
        value,
        bytes.fromhex(data[2:]),
        operation,
        safeTxGas,
        baseGas,
        gasPrice,
        gasToken,
        refundReceiver,
        nonce,
    ],
)

# === Domain separator ===
SAFE_ABI = [
    {
        "type": "function",
        "name": "getDomainSeparator",
        "inputs": [],
        "outputs": [{"type": "bytes32"}],
    }
]
contract = w3.eth.contract(SAFE_B, abi=SAFE_ABI)
domain_separator = contract.functions.getDomainSeparator().call()

# === Compute safeTxHash ===
inner_hash = keccak(packed)
safe_tx_hash = Web3.keccak(b"\x19\x01" + domain_separator + inner_hash).hex()

print("safeTxHash:", safe_tx_hash)
```

```python
from web3 import Web3
from eth_account import Account

###############################################################################
# CONFIG (EDIT THESE)
###############################################################################
RPC = "https://polygon-rpc.com"

SAFE_ADDRESS = Web3.to_checksum_address(TRADER_EOA_SAFE_MANUAL)
OWNER_PRIVATE_KEY = TRADER_EOA_PRIVATE_KEY
CHAIN_ID = 137  # Polygon

USDCe = Web3.to_checksum_address("0x2791Bca1f2de4661ED88A30C99A7a9449Aa84174")
MAX_UINT = (1 << 256) - 1

###############################################################################
# ERC20 ABI (only approve)
###############################################################################
ERC20_ABI = [
    {
        "inputs": [
            {"internalType": "address", "name": "spender", "type": "address"},
            {"internalType": "uint256", "name": "amount", "type": "uint256"},
        ],
        "name": "approve",
        "outputs": [{"internalType": "bool", "name": "", "type": "bool"}],
        "stateMutability": "nonpayable",
        "type": "function",
    }
]

###############################################################################
# Safe v1.3.0 ABI (execTransaction + getTransactionHash only)
###############################################################################
SAFE_ABI = [
    {
        "inputs": [
            {"internalType": "address", "name": "to", "type": "address"},
            {"internalType": "uint256", "name": "value", "type": "uint256"},
            {"internalType": "bytes", "name": "data", "type": "bytes"},
            {"internalType": "uint8", "name": "operation", "type": "uint8"},
            {"internalType": "uint256", "name": "safeTxGas", "type": "uint256"},
            {"internalType": "uint256", "name": "baseGas", "type": "uint256"},
            {"internalType": "uint256", "name": "gasPrice", "type": "uint256"},
            {"internalType": "address", "name": "gasToken", "type": "address"},
            {
                "internalType": "address",
                "name": "refundReceiver",
                "type": "address",
            },
            {"internalType": "bytes", "name": "signatures", "type": "bytes"},
        ],
        "name": "execTransaction",
        "outputs": [{"internalType": "bool", "name": "success", "type": "bool"}],
        "stateMutability": "payable",
        "type": "function",
    },
    {
        "inputs": [
            {"internalType": "address", "name": "to", "type": "address"},
            {"internalType": "uint256", "name": "value", "type": "uint256"},
            {"internalType": "bytes", "name": "data", "type": "bytes"},
            {"internalType": "uint8", "name": "operation", "type": "uint8"},
            {"internalType": "uint256", "name": "safeTxGas", "type": "uint256"},
            {"internalType": "uint256", "name": "baseGas", "type": "uint256"},
            {"internalType": "uint256", "name": "gasPrice", "type": "uint256"},
            {"internalType": "address", "name": "gasToken", "type": "address"},
            {
                "internalType": "address",
                "name": "refundReceiver",
                "type": "address",
            },
            {"internalType": "uint256", "name": "_nonce", "type": "uint256"},
        ],
        "name": "getTransactionHash",
        "outputs": [{"internalType": "bytes32", "name": "", "type": "bytes32"}],
        "stateMutability": "view",
        "type": "function",
    },
]

###############################################################################
# INIT
###############################################################################
w3 = Web3(Web3.HTTPProvider(RPC))
owner = Account.from_key(OWNER_PRIVATE_KEY)

token = w3.eth.contract(address=USDCe, abi=ERC20_ABI)
safe = w3.eth.contract(address=SAFE_ADDRESS, abi=SAFE_ABI)

###############################################################################
# 1. Create calldata: approve(SAFE, MAX_UINT)
###############################################################################
approve_data = token.encode_abi(
    "approve",
    args=[SAFE_ADDRESS, MAX_UINT],
)

print("Approve calldata:", approve_data)

###############################################################################
# 2. Compute Safe transaction hash
###############################################################################
safe_tx_hash = safe.functions.getTransactionHash(
    USDCe,  # to (USDC.e)
    0,  # value
    approve_data,  # data
    0,  # operation = CALL
    0,
    0,
    0,  # safeTxGas, baseGas, gasPrice
    "0x0000000000000000000000000000000000000000",  # gasToken
    "0x0000000000000000000000000000000000000000",  # refundReceiver
    safe.functions.nonce().call(),  # Safe nonce
).call()

print("Safe tx hash:", safe_tx_hash.hex())

###############################################################################
# 3. Sign Safe tx hash
###############################################################################
signature = owner.sign_hash(safe_tx_hash).signature
signatures = signature  # for 1-of-1 Safe; otherwise concat in owner order

print("Signature:", signatures.hex())

###############################################################################
# 4. Submit execTransaction
###############################################################################
tx = safe.functions.execTransaction(
    USDCe,
    0,
    approve_data,
    0,  # operation
    0,
    0,
    0,  # safeTxGas, baseGas, gasPrice
    "0x0000000000000000000000000000000000000000",
    "0x0000000000000000000000000000000000000000",
    signatures,
).build_transaction(
    {
        "from": owner.address,
        "nonce": w3.eth.get_transaction_count(owner.address),
        "gas": 300000,
        "gasPrice": w3.eth.gas_price,
    }
)

signed = owner.sign_transaction(tx)
tx_hash = w3.eth.send_raw_transaction(signed.rawTransaction)

print("Sent tx:", tx_hash.hex())
print("Polygonscan:", f"https://polygonscan.com/tx/{tx_hash.hex()}")
```
