```python
import os
from py_builder_relayer_client.client import RelayClient
from py_builder_signing_sdk.config import BuilderConfig, BuilderApiKeyCreds
import dotenv

dotenv.load_dotenv()

# relayer_url = os.getenv("RELAYER_URL")
relayer_url = "https://relayer-v2.polymarket.com/"
print(relayer_url)
chain_id = 137  # Polygon mainnet
private_key = os.getenv("QS_MASTER_EOA_PRIVATE_KEY")
# private_key = os.getenv("VALORY_TEST_PRIVATE_KEY")
# private_key = os.getenv("PRIVATE_KEY")
# private_key = os.getenv("TRADER_EOA_PRIVATE_KEY")

builder_api_key = os.getenv("BUILDER_API_KEY")
builder_secret = os.getenv("BUILDER_SECRET")
builder_passphrase = os.getenv("BUILDER_PASS_PHRASE")


# print(builder_api_key, "|", builder_secret, "|", builder_passphrase)

# Configure builder credentials
builder_config = BuilderConfig(
    local_builder_creds=BuilderApiKeyCreds(
        key=builder_api_key,
        secret=builder_secret,
        passphrase=builder_passphrase,
    )
)

# Initialize client
client = RelayClient(relayer_url, chain_id, private_key, builder_config)
```

```python
client.builder_config.signer.creds
```

```python
client.get_expected_safe()
```

```python
client.get_deployed(client.get_expected_safe())
```

```python
response = client.deploy()
result = response.wait()

if result:
    print("Safe deployed successfully!")
    print(f"Transaction Hash: {result.transaction_hash}")
    print(f"Safe Address: {result.proxy_address}")
else:
    print("Safe deployment failed")
```
